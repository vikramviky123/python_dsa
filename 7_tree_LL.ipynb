```python
class TreeNode:
    def __init__(self,data,children=[]):
        
        self.data = data
        self.children = children
        
    def __str__(self, level = 0):
        ret = " "*level + str(self.data) + "\n"
        
        for child in self.children:
            ret += child.__str__(level + 1)
            
        return ret
    
    def add_child(self,TreeNode):
        self.children.append(TreeNode)
```

```python
tree = TreeNode("Drinks", [])
cold = TreeNode("Cold", [])
hot = TreeNode("Hot", [])

tree.add_child(cold)
tree.add_child(hot)


tea = TreeNode("Tea", [])
coffee = TreeNode("Coffee", [])
hot.add_child(tea)
hot.add_child(coffee)

cola = TreeNode("Cola", [])
fanta = TreeNode("Fanta", [])
cold.add_child(cola)
cold.add_child(fanta)

print(tree)


```

**Binary Tree**
* ------> at most TWO Children

**Full Binary Tree**
* ------> Each node should have 0 or 2 child but NOT 1

**Perfect Binary Tree**
* ------> all nodes have two children....basically filled completely

**Complete Binary Tree**
* ------> all levels filled except last level, but last level should be filled left to right

**Balanced Binary Tree**
* ------> left sub tree - right sub tree should be 0 or 1

```python
class TreeNode:
    def __init__(self,data):
        self.data = data
        self.leftChild = None
        self.rightChild = None
```

```python
bt = TreeNode("Drinks")
leftChild = TreeNode("Hot")
rightChild = TreeNode("Cold")

tea = TreeNode("Tea")
coffee = TreeNode("Coffee")

cola = TreeNode("Cola")
fanta = TreeNode("Fanta")

bt.leftChild = leftChild
bt.rightChild = rightChild

leftChild.leftChild = tea
leftChild.rightChild = coffee

rightChild.leftChild = cola
rightChild.rightChild = fanta
```

```python
# level order traversal
# level wise data
import QueueLinkedList as queue
def levelOrderTraversal(rootNode):
    if not rootNode:
        return
    else:
        customQueue = queue.Queue()
        customQueue.enqueue(rootNode)
        while not(customQueue.isEmpty()):
            root = customQueue.dequeue()
            print(root.value.data)
            if (root.value.leftChild is not None):
                customQueue.enqueue(root.value.leftChild)
            
            if (root.value.rightChild is not None):
                customQueue.enqueue(root.value.rightChild)
                
    
levelOrderTraversal(bt)
```

```python
# search order
import QueueLinkedList as queue
def search_node(rootNode, nodeValue):
    if not rootNode:
        return
    else:
        customQueue = queue.Queue()
        customQueue.enqueue(rootNode)
        while not(customQueue.isEmpty()):
            root = customQueue.dequeue()
            if root.value.data == nodeValue:
                return "Success"
            if (root.value.leftChild is not None):
                customQueue.enqueue(root.value.leftChild)
            
            if (root.value.rightChild is not None):
                customQueue.enqueue(root.value.rightChild)
        
        return "Unsuccessful"
                
    
print(search_node(bt,"Tea"))
print(search_node(bt,"Milk"))
```

```python
# search order
import QueueLinkedList as queue
def insert_node(rootNode, newNode):
    if not rootNode:
        rootNode = newNode
    else:
        customQueue = queue.Queue()
        customQueue.enqueue(rootNode)
        while not(customQueue.isEmpty()):
            root = customQueue.dequeue()

            if (root.value.leftChild is not None):
                customQueue.enqueue(root.value.leftChild)
            else:
                root.value.leftChild = newNode
                return "Successfully inserted"
                
            
            if (root.value.rightChild is not None):
                customQueue.enqueue(root.value.rightChild)
            else:
                root.value.rightChild = newNode
                return "Successfully inserted"
        
      
                
newNode = TreeNode("TeaPowder")
print(insert_node(bt,newNode))
print(search_node(bt,"TeaPowder"))

```

The depth of a binary tree is the maximum number of edges from the root node to any leaf node.

```python
def depth(node):
    if node is None:
        return 0
    else:
        return max(depth(node.left), depth(node.right)) + 1
```

The height of a binary tree is the maximum number of edges from the root node to any leaf node. The height of an empty tree is -1, and the height of a binary tree having only one node is 0.

```python
# Deepest Node
def deepest_node(rootNode):
    if not rootNode:
        return
    else:
        customQueue = queue.Queue()
        customQueue.enqueue(rootNode)
        while not (customQueue.isEmpty()):
            root = customQueue.dequeue()

            if (root.value.leftChild is not None):
                customQueue.enqueue(root.value.leftChild)

            if (root.value.rightChild is not None):
                customQueue.enqueue(root.value.rightChild)

        deepestNode = root.value.data

        return deepestNode


print(deepest_node(bt))
```

```python
def height(root):
    if root is None:
        return -1
    maxi = height(root.leftChild)
    mini = height(root.rightChild)
    if mini<maxi:
        return maxi+1
    else:
        return mini+1
    
print(height(bt))
```
