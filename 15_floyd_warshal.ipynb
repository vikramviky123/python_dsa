#### Floyd Warshal
it is all pair short distance

```python
INF = 9999
# Printing the solution
def printSolution(nV, distance):
    for i in range(nV):
        for j in range(nV):
            if(distance[i][j] == INF):
                print("INF", end=" ")
            else:
                print(distance[i][j], end="  ")
        print(" ")


def floydWarshall(nV, matrix):
    distance = matrix.copy()
    for k in range(nV):
        for i in range(nV):
            for j in range(nV):
                print(f"i = {i} & j = {j} is ",distance[i][j], f"\ni = {i} & k = {k} is ",distance[i][k],f"\nk = {k} & j = {j} is ",distance[k][j])
                distance[i][j] = min(distance[i][j], distance[i][k]+distance[k][j])
                print("------")
            print(distance)
            print("-"*80)
    
    printSolution(nV, distance)

```

```python
G = [[0, 8, INF,1],
    [INF, 0, 1,INF],
    [4, INF, 0,INF],
    [INF, 2, 9,1]
    ]
GG = [[0, 3, INF, 5],
     [2, 0, INF, 4],
     [INF, 1, 0, INF],
     [INF, INF, 2, 0]]
```

```python
floydWarshall(4, GG)
```
