### Bubble Sort
* for first full loop the large value is fully sorted

```python
def bubble_sort(customList):
    count = 0
    
    for i in range(len(customList)-1):
        for j in range(len(customList)-i-1):
            count+=1
            if customList[j] > customList[j+1]:
                customList[j] ,customList[j+1] = customList[j+1],customList[j] 
                
    print(customList,"\n",count,"\n",count/len(customList))
    
lst = [9,8,7,6,5,4,3,2,1]
bubble_sort(lst)          
```

### Selection Sort
* search for min element and swap with first element

```python
def selection_sort(customList):
    count = 0
    
    for i in range(len(customList)):
        min_index = i
        for j in range(i+1,len(customList)):
            count+=1
            if customList[min_index] > customList[j]:
                min_index = j        
        customList[i], customList[min_index] = customList[min_index], customList[i]
       
    print(customList,"\n",count,"\n",count/len(customList))

lst = [9,8,7,6,5,4,3,2,1]
selection_sort(lst)   
    
```

### Insertion Sort
* sorted array and unsorted array

```python
def insertion_sort(customList):
    count = 0
    for i in range(1,len(customList)):
        key = customList[i]
        j = i-1
        
        while j >= 0 and key < customList[j]:
            count+=1
            customList[j+1] = customList[j]
            j-=1
            
        customList[j+1] = key
        
    return print(customList,"\n",count)

lst = [9,8,7,6,5,4,3,2,1]
insertion_sort(lst)  
```

### Bucket Sort
* vaule x no_of_buckets/length_of List
* no_of_buckets = sqrt(length_of)

```python
import math
def bucket_sort(customList):
    length_of_list = len(customList)
    no_of_buckets = round(math.sqrt(length_of_list))
    maxValue = max(customList)
    
    arr = []
    
    for i in range(no_of_buckets):
        arr.append([])
        
    for j in customList:
        index_b = math.ceil(j*no_of_buckets/maxValue)
        arr[index_b-1].append(j)
        
    for i in range(no_of_buckets):
        arr[i] = insertion_sort(arr[i])
        
    k = 0
    
    for i in range(no_of_buckets):
        for j in range(len(arr[i])):
            customList[k] = arr[i][j]
            k+=1
    return customList

lst = [9,8,7,6,5,4,3,2,1]
bucket_sort(lst)      
```

### Merge Sort
![Alt text](image-5.png)

```python
def merge(customList, l, m, r):
    n1 = m - l + 1
    n2 = r - m

    L = [0] * (n1)
    R = [0] * (n2)

    for i in range(0, n1):
        L[i] = customList[l+i]
    
    for j in range(0, n2):
        R[j] = customList[m+1+j]
    
    i = 0 
    j = 0
    k = l
    while i < n1 and j < n2:
        if L[i] <= R[j]:
            customList[k] = L[i]
            i += 1
        else:
            customList[k] = R[j]
            j += 1
        k += 1
    while i < n1:
        customList[k] = L[i]
        i += 1
        k += 1
    
    while j < n2:
        customList[k] = R[j]
        j += 1
        k += 1
        
def merge_sort(customList,l,r):
    
    if l < r:
        m = (l+(r-1))//2
        merge_sort(customList,l,m)
        merge_sort(customList,m+1,r)
        merge(customList,l,m,r)
        
    return customList
    
lst = [9,8,7,6,5,4,3,2,1]
print(merge_sort(lst,0,8))
```

### Quick Sort
![Alt text](image-6.png)

```python
def quick_sort(arr, left, right):
    if left < right:
        partition_pos = partition(arr,left,right)
        quick_sort(arr,left,partition_pos-1)
        quick_sort(arr,partition_pos+1, right)

def partition(arr,left,right):
    i = left
    j = right - 1
    pivot = arr[right]
    
    while i < j:
        while i < right and arr[i] < pivot:
            i += 1
        while j > left and arr[j] >= pivot:
            j -= 1
            
        if i < j:
            arr[i], arr[j] = arr[j], arr[i]
            
    if arr[i] > pivot:
        arr[i], arr[right] = arr[right], arr[i]
        
    return i

arr = [22,11,88,66,55,77,33,44]
arr_ = [22,11,33,44,55,77,88,66]
quick_sort(arr,0,len(arr) - 1)
print(arr)
         
     
```

### Heap Sort

```python
def heapify(customList, n, i):
    smallest = i
    l = 2*i + 1
    r = 2*i + 2
    
    if l < n and customList[l] < customList[smallest]:
        smallest = l
    
    if r < n and customList[r] < customList[smallest]:
        smallest = r
    
    if smallest != i:
        customList[i], customList[smallest] = customList[smallest], customList[i]
        heapify(customList, n, smallest)


def heapSort(customList):
    n = len(customList)
    mid = int(n/2)-1
    end = n - 1
    
    for i in range(mid, -1, -1):
        heapify(customList, n, i)
    
    for i in range(end,0,-1):
        customList[i], customList[0] = customList[0], customList[i]
        heapify(customList, i, 0)
    customList.reverse()
    
arr = [2,1,7,6,5,3,4,9,8]
heapSort(arr)
print(arr)
```

### Binary Search

```python
import math
def binarySearch(array, value):
    start = 0
    end = len(array)-1
    middle = math.floor((start+end)/2)
    
    while (array[middle] != value) and start <= end:
        if value < array[middle]:
            end = middle - 1
        else:
            start = middle + 1 
        middle = math.floor((start+end)/2)
        # print(start, middle, end)
    if array[middle] == value:
        return middle
    else:
        return -1     

custArray = [8, 9, 12, 15, 17, 19, 20, 21, 28]
print(binarySearch(custArray, 15))
```
